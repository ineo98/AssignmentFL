```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
# Create variebles for antecedent(conditions) and consequent(conclusion) 
red = ctrl.Antecedent(np.arange(0,256,0.01), 'red')
green = ctrl.Antecedent(np.arange(0,256,0.01), 'green')
blue = ctrl.Antecedent(np.arange(0,256,0.01), 'blue')

ripeness = ctrl.Consequent(np.arange(0, 91, 1), 'ripeness')
```

```python
# Create membership function for antecedent-red
red['low'] = fuzz.trapmf(red.universe, [0, 0, 90.33, 142.38])
red['medium'] = fuzz.trimf(red.universe, [90.33, 142.38, 176.18])
red['high'] = fuzz.trapmf(red.universe, [142.38, 176.18, 255, 255])

red.view()
```

```python
# Create membership function for antecedent-green
green['low'] = fuzz.trapmf(green.universe, [0, 0, 105.64, 130.64])
green['medium'] = fuzz.trimf(green.universe, [105.64, 130.64, 147.44])
green['high'] = fuzz.trapmf(green.universe, [130.64, 147.44, 255, 255])

green.view()
```

```python
# Create membership function for antecedent-green
blue['low'] = fuzz.trapmf(blue.universe, [0, 0, 27.10, 39.84])
blue['high'] = fuzz.trapmf(blue.universe, [27.10, 39.84, 255, 255])

blue.view()
```

```python
# Create membership function for antecedent-green
ripeness['unripe'] = fuzz.trapmf(ripeness.universe, [0, 0, 15, 45])
ripeness['underripe'] = fuzz.trimf(ripeness.universe, [15, 45, 75])
ripeness['ripe'] = fuzz.trapmf(ripeness.universe, [45, 75, 90, 90])

ripeness.view()
```

**Rule base**

```python
rule1 = ctrl.Rule(red['low'] & green['low'], ripeness['unripe'])
rule2 = ctrl.Rule(red['low'] & green['medium'], ripeness['unripe'])

rule3 = ctrl.Rule(red['medium'] & green['medium'], ripeness['underripe'])
rule4 = ctrl.Rule(red['medium'] & green['high'], ripeness['underripe'])

rule5 = ctrl.Rule(red['high'] & green['medium'], ripeness['ripe'])
rule6 = ctrl.Rule(red['high'] & green['high'], ripeness['ripe'])
```

```python
Cntrl_sys = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])

Ripeness = ctrl.ControlSystemSimulation(Cntrl_sys)

Ripeness.input['red'] = 170.4637
Ripeness.input['green'] = 135.1407

Ripeness.compute()
```

```python
print(Ripeness.output['ripeness'])
```

```python
ripeness.view(sim=Ripeness)
```
